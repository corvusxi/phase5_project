# Train Test Split

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=53, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Models: Random Forest

```python
rfsteps = [("smote", SMOTE(random_state=53)),
         ("scaler", StandardScaler()),
         ("rf", RandomForestClassifier(random_state=53))]

rfpipe = Pipeline(rfsteps)
```

```python
rfpipe.fit(X_train, y_train); #7m run time
```

```python
y_rfpred = rfpipe.predict(X_test)
```

```python
print(classification_report(y_test, y_rfpred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_rfpred,
                                        display_labels=["0: Not Fraudulent", "1: Fraudulent"],
                                        colorbar=True, cmap="YlGn")
plt.grid(False);
```

100% accuracy, 80% recall on type 1. This model is noticably good at True Negatives in comparison to False Positives, but it is concerning that False Negatives are higher than False Positives.

Seeing what some tuning can adjust.

```python
RocCurveDisplay.from_estimator(rfpipe, X_test, y_test)
plt.plot([0, 1], [0, 1], "k--");
```
