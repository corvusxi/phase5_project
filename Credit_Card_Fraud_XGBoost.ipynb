```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

from sklearn.metrics import classification_report, ConfusionMatrixDisplay, RocCurveDisplay
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler

from xgboost import XGBClassifier
```

# Train Test Split

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=53, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Models: XGBoost

```python
xgbsteps = [("smote", SMOTE(random_state=53)),
         ("scaler", StandardScaler()),
         ("xgb", XGBClassifier(objective="reg:squarederror"))]

xgbpipe = Pipeline(xgbsteps)
```

```python
xgbpipe.fit(X_train, y_train)
```

```python
y_pred_xgbpipe = xgbpipe.predict(X_test)
print(classification_report(y_test, y_pred_xgbpipe))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_xgbpipe,
                                        display_labels=["0: Not Fraudulent", "1: Fraudulent"])
plt.grid(False);
```

```python
RocCurveDisplay.from_estimator(xgbpipe, X_test, y_test)
plt.plot([0, 1], [0, 1], "k--");
```

```python
xgbparams = {"xgb__max_depth": [3, 4, 5], "xgb__learning_rate": [.1, .25, .5], "xgb__reg_alpha": [0, .25, .5], "xgb__reg_lambda": [0, .25, .5],
             "xgb__n_estimators": [100, 150, 200]}

xgbgrid = GridSearchCV(estimator=xgbpipe, param_grid=xgbparams, cv=5, scoring="recall") #62m run time
```

```python
xgbgrid.fit(X_train, y_train);
```

```python
xgbgrid.best_params_
```

```python
xgbgrid.best_estimator_
```

```python
xgbgrid.best_score_
```

```python
bestxgbgrid = xgbgrid.best_estimator_
bestxgbgrid.fit(X_train, y_train);
y_xgbpred = bestxgbgrid.predict(X_test)
```

```python
y_xgbpred = bestxgbgrid.predict(X_test)
print(classification_report(y_test, y_xgbpred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_xgbpred, display_labels=["0: Not Fraudulent", "1: Fraudulent"])
plt.grid(False);
```
