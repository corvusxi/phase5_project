# Train Test Split

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=53, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# SVM

Very new to SVM so this will be rather messy!

```python
svm = OneClassSVM(kernel="poly", nu=0.05, gamma=0.07, degree=3, coef0=0.02)
```

```python
svm.fit(X_train)
```

```python
predict = svm.predict(X_test)
predict = [1 if i==-1 else 0 for i in predict]
```

```python
print(classification_report(y_test, predict))
```

```python
score = svm.score_samples(X_test)
```

```python
score_threshold = np.percentile(score, 2)
print(f"The customized score threshold for 2% of outliers is {score_threshold:.2f}")
```

```python
custom_predict  = [1 if i < score_threshold else 0 for i in score]
```

```python
print(classification_report(y_test, custom_predict))
```

```python
dfvis = pd.DataFrame(X_test, columns=["V4", "V5", "V8"])
dfvis["y_test"] = y_test
dfvis["svm prediction"] = predict
dfvis["svm custom prediction"] = custom_predict
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(1,3, sharey=True)

#Ground
ax0.set_title("Original")
ax0.scatter(dfvis["V4"], dfvis["V5"], dfvis["V8"], c=dfvis["y_test"], cmap="summer");

ax1.set_title("One Class SVM Predictions")
ax1.scatter(dfvis["V4"], dfvis["V5"], dfvis["V8"], c=dfvis["svm prediction"], cmap="summer");

ax2.set_title("One Class SVM Predictions")
ax2.scatter(dfvis["V4"], dfvis["V5"], dfvis["V8"], c=dfvis["svm custom prediction"], cmap="summer");
```

# SVM with Pipeline

```python
svmsteps = [("smote", SMOTE(random_state=53)),
            ("scaler", StandardScaler()),
            ("svm", OneClassSVM(kernel="poly", nu=0.05, gamma=0.07, degree=3, coef0=0.02))]

svmpipe = Pipeline(svmsteps)
```
