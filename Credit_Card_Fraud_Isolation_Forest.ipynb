```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, roc_curve, RocCurveDisplay, make_scorer, recall_score
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
```

# Train Test Split

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=53, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Models: Isolation Forest

```python
ifsteps = [("smote", SMOTE(random_state=53)),
         ("scaler", StandardScaler()),
         ("if", IsolationForest(random_state=53))]

ifpipe = Pipeline(ifsteps)
```

```python
ifpipe.fit(X_train, y_train);
```

```python
y_ifpred = ifpipe.predict(X_test)
y_pred_binary = [1 if label == -1 else 0 for label in y_ifpred]
print(classification_report(y_test, y_pred_binary))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_binary, display_labels=["0: Not Fraudulent", "1: Fraudulent"])
plt.grid(False);
```

Tuning

```python
ifparams = {"if__n_estimators": [50, 100, 200],
            "if__max_features": range(1,30)}

ifgrid = GridSearchCV(estimator=ifpipe, param_grid=ifparams, cv=5, scoring="recall")
```

```python
ifgrid.fit(X_train, y_train)
```
