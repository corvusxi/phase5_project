# Train Test Split

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=53, stratify=y)
```

```python
Xnp = X.iloc[:,1:].values
ynp = y.to_numpy()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Models: Isolation Forest

```python
ifsteps = [("smote", SMOTE(random_state=53)),
         ("scaler", StandardScaler()),
         ("if", IsolationForest(random_state=53, max_samples=100, contamination=0.2))]

ifpipe = Pipeline(ifsteps)
```

```python
ifpipe.fit(X_train, y_train);
```

```python
df["Anomaly Scores"] = ifpipe.decision_function(X_train, y_train)
```

```python
y_ifpred = ifpipe.predict(X_test)
y_pred_binary = [1 if label == -1 else 0 for label in y_ifpred]
print(classification_report(y_test, y_pred_binary))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred_binary, display_labels=["0: Not Fraudulent", "1: Fraudulent"],
                                        colorbar=True, cmap="YlGn")
plt.grid(False);
```

Baseline Isolation Forest is doing very poorly on this with a lot of concern regarding the False Negatives.
