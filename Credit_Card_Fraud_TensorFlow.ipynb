# Train Test Split

```python
X = df.drop(columns=["Class"])
y = df["Class"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=53, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Model

```python
trainCallback = EarlyStopping(monitor="loss", min_delta=1e-6, patience=5)
```

Creating a model for the Pipeline:

```python
def tfmodel():
    model = Sequential()
    model.add(Dense(10000, input_dim=30, activation="relu", kernel_regularizer=L2(.01)))
    model.add(Dense(10, activation="relu", kernel_regularizer=L1(.005)))
    model.add(Dense(units=1, activation="sigmoid"))
    optimizer = Adam(learning_rate=0.01, beta_1=0.9, beta_2=0.999)
    model.compile(loss="binary_crossentropy", optimizer=optimizer, metrics=["accuracy", "AUC"])
    return model
```

```python
tfclass = KerasClassifier(build_fn=tfmodel, epochs=100, batch_size=32, verbose=1, callbacks=[trainCallback])
```

# TensorFlow

```python
tfsteps = [("smote", SMOTE(random_state=53)),
         ("scaler", StandardScaler()),
         ("tf", tfclass)]

tfpipe = Pipeline(tfsteps)
```

```python
tfpipe.fit(X_train, y_train)
```

```python
y_tfpred = tfpipe.predict(X_test)
print(classification_report(y_test, y_tfpred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_tfpred,
                                        display_labels=["0: Not Fraudulent", "1: Fraudulent"],
                                        colorbar=True, cmap="YlGn")
plt.grid(False);
```

Very similar to Gradient Boost, with a few of True Positives moved into False Positives.

```python
RocCurveDisplay.from_estimator(tfpipe, X_test, y_test)
plt.plot([0, 1], [0, 1], "k--");
```
